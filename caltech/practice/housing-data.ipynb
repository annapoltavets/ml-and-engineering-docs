```python
df_one_hot = pd.get_dummies(df['Neighborhood'], sparse=True)
```

```python
df.groupby(['Neighborhood'])['SalePrice'].mean()
```

```python
df11 = df[['Neighborhood', 'SalePrice']]
```

```python
df2 = pd.concat([df11, df_one_hot], axis=1)
df2
```

```python
def prices(n, p):
    if n == False:
        return None
    else:
        return p
    
```

```python
for c in df_one_hot.columns:
    df2[c] = df2.apply(lambda row: prices(row[c], row['SalePrice']), axis=1)
```

```python
df4 = df3.T
df5 = df4[df4.index != 'SalePrice']
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
plt.xticks(rotation=90)
sns.lineplot(df5['count'])
```

```python
df6 = df5[df5['count'] >= 25]
```

```python
import seaborn as sns
sns.choose_dark_palette()
plt.xticks(rotation=90)

sns.lineplot(x=df6.index, y='25%', data=df6, label="Q1", color="yellow")
sns.lineplot(x=df6.index, y='50%', data=df6, label="Q2", color="orange")
sns.lineplot(x=df6.index, y='75%', data=df6, label="Q3", color="red")
sns.lineplot(x=df6.index, y='mean', data=df6, label="mean", color="blue")
plt.xlabel("Neighborhood")
plt.ylabel("Price")
```
