# Sales Forecast

## Problem statement

In the ever-changing competitive market conditions, there is a need to make correct
business decisions and plan for future events. The effectiveness of a business decision is
influenced by the accuracy of the models used. Demand is the most important aspect of
a business's ability to achieve its objectives. Many decisions in business depend on
demand, like production, sales, and staff requirements. Forecasting is necessary for
business at both international and domestic levels.

## Problem objective
Fresh Analytics, a data analytics company, aims to comprehend and predict the demand
for various items across restaurants. To accurately forecast the sales of items across
different restaurants over the years, enabling better decision-making and planning.

### Load data set

```python
import pandas as pd
```

```python
items = pd.read_csv('data/items.csv', sep=',')
print(items.shape)
resturants = pd.read_csv('data/resturants.csv', sep=',')
print(resturants.shape)
sales = pd.read_csv('data/sales.csv', sep=',')
print(sales.shape)
merged_df = sales.merge(items, left_on='item_id', right_on='id', how='inner', suffixes=('', '_item')).drop('id', axis=1)
merged_df = merged_df.merge(resturants, left_on='store_id', right_on='id', how='inner', suffixes=('', '_store')).drop('id', axis=1)
merged_df['sales'] = merged_df['price'] * merged_df['item_count']
print(merged_df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def show_plots_board(pdf, cols, axes, annotate=True, plot='histplot'):
    axes = axes.flatten()
    for i, col in enumerate(cols):
        if i < len(axes):
            if pdf[col].notna().any():
                if plot == 'boxplot':
                    sns.boxplot(pdf[col], ax=axes[i])
                else:
                    sns.histplot(pdf[col].dropna(), kde=True, ax=axes[i])

                axes[i].set_title(f'Distribution of {col}')

                if annotate:
                    # Add skewness and kurtosis
                    skewness = pdf[col].skew()
                    kurtosis = pdf[col].kurt()
                    axes[i].annotate(f'Skewness: {skewness:.2f}\nKurtosis: {kurtosis:.2f}',
                                     xy=(0.7, 0.85), xycoords='axes fraction')

    plt.tight_layout()
    plt.show()


# Visualize distributions of key numerical variables
fig, axes = plt.subplots(1, 6, figsize=(15, 3))
cols = ['item_id', 'price', 'item_count', 'store_id', 'kcal', 'cost']
show_plots_board(merged_df, cols, axes, False, 'boxplot')
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))
show_plots_board(merged_df, cols, axes, False, 'histplot')
```

### Generate necessary features for the development of these models, like day of the week, quarter of the year, month, year, day of the month and so on

```python
import holidays as hdays
df = merged_df.copy().sort_values(by='date')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date_id'] = df['date'].rank(method='dense').astype(int)

s = df['date']
df["year"]         = s.dt.year
df["quarter"]      = s.dt.quarter
df["month"]        = s.dt.month
df["day"]          = s.dt.day
df["day_of_year"]  = s.dt.dayofyear
df["weekday"]      = s.dt.weekday

years = pd.Series(df["year"]).dropna().unique().tolist()
holidays = set(hdays.country_holidays("US", years=years).keys())
df["is_holiday"]   = s.dt.date.isin(holidays).astype('int')
df["is_weekend"]   = df["weekday"].isin([7,0]).astype('int')
df["is_business_day"] = ((df["is_weekend"] == 0) & (df["is_holiday"] == 0)).astype('int')
```

### Build and compare linear regression, random forest, and XGBoost models for predictions
 Use the data from the last six months as the testing data
 Compute the root mean square error (RMSE) values for each model to compare their performances
 Use the best-performing models to make a forecast for the next year

```python
import time
import numpy as np

from sklearn.metrics import mean_squared_error

class Stats:
    def __init__(self, model, model_name):
        self.model_name = model_name
        self.model = model
        self.train_time = 0
        self.inference_time = 0
        self.rmse = 0
        self.test_score = 0
        
    def train_model(self, X_train, y_train):
        t1 = time.time()
        self.model.fit(X_train, y_train)
        t2 = time.time()
        self.train_time = t2 - t1
        print(f"Training time: {self.train_time:.2f} seconds")

    def predict(self, X):
        t0 = time.time()
        y_pred = self.model.predict(X)
        t1 = time.time()
        self.inference_time = t1 - t0
        print(f"Inference time: {self.inference_time:.2f} seconds")
        return y_pred

    def calculate_rmse(self, y_true, y_pred):
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        return rmse

    def calculate_accuracy(self, X_test, y_test):
        self.test_score = self.model.score(X_test, y_test)
        print(f"Test score: {self.model.score(X_test, y_test)}")
        y_pred = self.predict(X_test)
        rmse = self.calculate_rmse(y_test, y_pred)
        self.rmse = rmse
        print(f"RMSE: {rmse:.2f}")

    def fit(self, X_train, y_train, X_test, y_test):
        print(f"\nTraining {self.model_name}...")
        self.train_model(X_train, y_train)
        self.calculate_accuracy(X_test, y_test)

    def __str__(self):
        return f"Train time: {self.train_time:.2f}s, Inference time: {self.inference_time:.2f}s, RMSE: {self.rmse:.2f}"
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

train_df = df[df['date'] <= (df['date'].max() - pd.Timedelta(days=180))]
test_df = df[df['date'] > (df['date'].max() - pd.Timedelta(days=180))]
X_train = train_df.drop(columns=['sales', 'date', 'date_id','name', 'name_store'])
y_train = train_df['sales']
X_test = test_df.drop(columns=['sales', 'date', 'date_id', 'name', 'name_store'])
y_test = test_df['sales']

stats = []

s = Stats(LinearRegression(), 'Linear Regression')
s.fit(X_train, y_train, X_test, y_test)
stats.append(s)

s = Stats(RandomForestRegressor(n_estimators=20), 'Random Forest Regressor')
s.fit(X_train, y_train, X_test, y_test)
stats.append(s)

s = Stats(RandomForestRegressor(n_estimators=20), 'Random Forest Regressor')
s.fit(X_train, y_train, X_test, y_test)
stats.append(s)


s = Stats(GradientBoostingRegressor(n_estimators=200, learning_rate=0.001, max_depth=3, random_state=42), 'Gradient Boosting Regressor 200')
s.fit(X_train, y_train, X_test, y_test)
stats.append(s)

s = Stats(GradientBoostingRegressor(n_estimators=1000, learning_rate=0.001, max_depth=3, random_state=42), 'Gradient Boosting Regressor 1000')
s.fit(X_train, y_train, X_test, y_test)
stats.append(s)
```
