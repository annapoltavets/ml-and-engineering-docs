```python
import pandas as pd

def get_id_by_name(movie_name):
    try:
        data = requests.get(f"{BASE_URL}/search/movie", params={"query": movie_name}, headers=HEADERS).json()
        ids = [x['id'] for x in data['results']]
        return ids
    except IndexError:
        print(f'Cannot find: {movie_name}')

def find_csv_by_imdb(ids, imdb):
    for id in ids:
        movie = requests.get(f"{BASE_URL}/movie/{id}", headers=HEADERS).json()
        if 'imdb_id' in movie and movie['imdb_id'] == imdb:
            original_language = movie['original_language']
            overview = movie['overview']
            popularity = movie['popularity']
            release_date = movie['release_date']
            vote_average = movie['vote_average']
            vote_count = movie['vote_count']
            budget = movie['budget']
            revenue = movie['revenue']

            production_companies_names = '|'.join([x['name'] for x in movie['production_companies']])
            production_companies_id = '|'.join([str(x['id']) for x in movie['production_companies']])
            production_companies_country = '|'.join([x['origin_country'] for x in movie['production_companies']])

            m = f'{id}\t{imdb}\t{original_language}\t{overview}\t{popularity}\t{release_date}\t{vote_average}\t{vote_count}\t{production_companies_names}\t{production_companies_id}\t{production_companies_country}\t{budget}\t{revenue}'
            return m



#2000, 2001,
movie_list = []
years = [2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

for year in years:
    imdb_titles = pd.read_csv(f'data/imdb/titles_{year}.csv')
    for index, row in imdb_titles.iterrows():
        print(row['tconst'])
        ids = get_id_by_name(row['title'])
        result = find_csv_by_imdb(ids, row['tconst'])
        if result != None:
            movie_list.append(result)
        
        if index % 50 == 0 and index > 0:
            print(f'Processed {index} movies')
            with open('data/imdb/movies2000_2020.tsv', 'a', encoding='utf-8') as f:
                f.write('\n'.join(movie_list))
            movie_list = []
```
