```python
import pandas as pd
from imdb_base_data_loader import ImdbBaseDataLoader
from imdb_principals_data_loader import ImdbPrincipalDataLoader
dl = ImdbBaseDataLoader(from_cache=True)
pdl = ImdbPrincipalDataLoader(dl.imdb_base, from_cache=False)
p = pdl.principals
```

```python
tmp = p[(p['title_type'] == 'movie') & (p['start_year'] >= 1985) &  (p['start_year'] < 1995) & (p['imdb_num_votes']  > 1000)]

#

(tmp[tmp['nconst_name'].isin(['Steve Guttenberg',
                 'Shelley Long',
                 'Jami Gertz',
                 'Kyle MacLachlan',
                  'Dustin Hoffman',
                  'Will Ferrell',
                  'Bill Murray',
                  'Matthew Broderick',
                  'Frank Whaley',
                  'Sharon Stone',
                  'Michael Douglas',
                  'Matt Damon',
                  'Melanie Griffith',
                  'Joan Cusack', 
                  'Goldie Hawn',
                  'Kurt Russell',
                  'Harrison Ford',
                  'Jim Belushi', 'Tom Hanks'
                              ])][['imdb_rating', 'imdb_num_votes', 'title_type', 'title', 'start_year', 'genres', 'nconst_name' ]]
    .sort_values(by=['imdb_rating', 'imdb_num_votes'], ascending=[True, True])

)
# Falling Down, Witness, Broadcast News, Wall Street, Big Trouble in Little China, School Ties, Something Wild, Presumed Innocent, Clear and Present Danger, Patriot Games, The War of the Roses, Paradise, Hoffa, Men Don't Leave, Pacific Heights
```
