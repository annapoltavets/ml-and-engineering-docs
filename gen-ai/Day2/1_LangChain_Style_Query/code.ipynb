# Langchain New Style ----- from 0.3 onwards 

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
result = llm.invoke("Tell me a joke")
```

```python
print(result.content)
```

```python
# SystemMessage ---- help to setup persona of the LLM
# HumanMessage  ---- user input | Input Suppled from human
# AIMessage     ---- Response generated by the LLM (Assistant)


from langchain.schema import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage(content="Act as a Data Engineer who is an expert in Pandas package and help solve problems related to Pandas"),
    HumanMessage(content="How to create DataFrame from a csv file?")

]

result = llm.invoke(input=messages)
```

```python
print(result.content)
```

```python
#Batch Message

batchMessages = [
    [
        SystemMessage(content="Act as Howie Mandel"),
        HumanMessage(content="Tell me a joke")
    ],
    [
        SystemMessage(content="Act as a dietician"),
        HumanMessage(content="Suggest me Indian Keto diet plan for Non-veg user")
    ]
]
```

```python
result = llm.generate(batchMessages)
```
